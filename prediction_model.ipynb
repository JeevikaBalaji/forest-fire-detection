```python
#Importing the necessary libraries
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
import librosa
from tensorflow.keras.preprocessing import image
from keras.models import load_model
```

```python
video = cv2.VideoCapture('/content/Wildfires 101 _ National Geographic - Trim.mp4')
```

```python
# Check if camera opened successfully
if (video.isOpened()== False):
  print("Error opening video stream or file")

# Read until video is completed
while(video.isOpened()):
  # video frame-by-frame
  ret, frame = video.read()
  if ret == True:
    x=image.img_to_array(frame)
    res=cv2.resize(x,dsize=(128,128),interpolation=cv2.INTER_CUBIC)
    #expand the image shape
    x=np.expand_dims(res,axis=0)
    model=load_model("/content/drive/MyDrive/forest fire detection../forest1.h5")
    cv2_imshow(frame)
    pred=model.predict(x)
    pred = int(pred[0][0])
    pred
    int(pred)
    if pred==0:
      print('Forest fire detected')
      break
    else:
      print("no danger")
      break

# When everything done, release the video video object
video.release()

# Closes all the frames
cv2.destroyAllWindows()
```
